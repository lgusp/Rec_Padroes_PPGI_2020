-- Vem com uma linha de idxxx e a outra com as colunas - vamos juntar as duas acertando o id na primeira coluna
-- Verificação na quantidade final de colunas - x colunas
-- No arquivo original os valores estão separados por ponto-e-vírgula e a cada 4 valor a separação é só por vírgula
-- Vamos deixar tudo como virgula

```python
# Rotina de acerto de arquivo
def ajustaArquivo(arqent, arqsai):
    import re
    
    arqE = arqent # Arquivo de Entrada
    arqS = arqsai # Arquivo de saída
    print('\nArquivos Tratados. Entrada e saída: ', arqE, arqS)

    # Tratando o identificador da linha
    linha = ''
    lnId = ''
    lnVal = ''
    ls = ''
    linhas = []

    with open(arqS, 'w') as saida:
        with open(arqE) as entrada:
        
            for le in entrada:
                #print('Linha de Entrada: ', le)         
                
                # Retira os espaços em branco da linha (antes e depois)
                linha = le.rstrip()

                # Despreza as linhas em branco e a última linha
                if (linha == '') or (linha[0:17] == 'Descriptor SPHARM'):
                    continue
                    
                # Trata das linhas com o identificador do paciente
                if (linha[0:8] == 'Paciente'):
                    if (len(linha) > 13) or (len(linha) < 2):
                        print('**** Erro. Linha errada: ', linha)
                    elif (len(linha) == 11):
                        lnId = 'id00' + linha[10:11]
                    elif (len(linha) == 12):
                        lnId = 'id0' + linha[10:12]
                    else:
                        lnId = 'id' + linha[10:13]
                    continue
                
                # Trata das linhas complementares
                if (linha[0:5] == 'VALOR'):
                    lnVal = linha
                    continue
                    
                # Precisa ser uma linha de dados numéricos
                # Junta o id e valor arrumado com a linha de dados
                if (lnVal != ''):
                    ls = lnId + ',' + lnVal + ',' + linha[:-1]
                elif (lnId != ''):
                    ls = lnId + ',' + linha[:-2]
                else:
                    print('****Erro:', linha)
                
                           
                    
                # Altera todas os ponto-e-vírgulas para vírgulas
                ls = ls.replace(";", ",")
                                    
                # Grava a linha arrumada com o id e os dados no arquivo de saída
                linhas.append(ls)
                #print('Linha gravada: ', ls)
                
                linha = ''
                lnVal = ''
                
        # Para cada ocorrência em linha, verifica a quantidade de virgulas
        lmax = 0
        for li in linhas:
            if (li.count(',')) > lmax:
                lmax = li.count(',')
                
        print('Quantidade de Virgulas:', lmax)
    
        # Com a quantidade de virgulas máxima, iguala todas as linhas pra esta quantidade
        for li in linhas:
            difv = lmax - li.count(',')

            for i in range(difv):
                li = li + ','
            
            li = li + '\n'
        
            # Grava a linha arrumada com o id e os dados no arquivo de saída
            saida.write(li)
            #print('Linha gravada: ', linha)
                
    
    # Verificando o arquivo gravado
    df = pd.read_csv(arqsai,  header=None)
    print('\n*** Dados do DataFrame ***')
    print('**** Saída ****')
    print('**************************')
    print(f'Linhas e Colunas: {df.shape}')
    print()
    print(df.head())
    print(df.tail())
    
```
