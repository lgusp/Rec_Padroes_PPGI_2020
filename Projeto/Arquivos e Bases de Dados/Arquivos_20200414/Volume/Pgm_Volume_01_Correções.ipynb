## Tratamento de Arquivo
### Arquivo de Entrada: volume_tratado_Excel_20200414.csv

#### Tratamentos:
- A primeira linha é o cabeçalho e nomes das colunas no DataFrame
- Somente até a linha 401 (cabeçalho mais 400 amostras)
- Ajuste do identificador do paciente - será idxxx
- Alteração da informação genêro para 0(M) e 1(F)
- Colunas: ID, IDADE, SEXO

### Arquivo de Saída: Volume_20200414.csv

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('volume_tratado Excel_20200414.txt')

print('*** Dados do DataFrame ***')
print('**************************')
print(f'Linhas e Colunas: {df.shape}')
print()
print(df.info())
print()
df.head()
```

```python
# Ajustando o padrão do id do paciente - coluna 1, incluindo o idxxx

print('**** Convertendo para inteiro ****')
df = df.astype({"ID": str})

for l in range(len(df)):
    dx = df.iloc[l, 0]
    
    if (len(dx) == 1):
        df.iloc[l, 0] = 'id00' + dx[0:]
    elif (len(dx) == 2):
        df.iloc[l, 0] = 'id0' + dx[0:]
    else:
        df.iloc[l, 0] = 'id' + dx[0:]
        
print('*** Dados do DataFrame ***')
print('**************************')
print(f'Linhas e Colunas: {df.shape}')
print()
print(df.info())
print()
print(df.head())
print()
print(df.tail())
```

```python
# Tratando os dados faltantes de Volume e Massa - média das colunas
print('**** Nulos ou NaN ****')
dfm = df['Volume'].isnull()
print(df[dfm])
print()

print('**** Colocando a média onde houver NaN ****')
media = df['Volume'].mean()

df.loc[dfm, 'Volume'] = media
print(df[dfm])
print()

print('**** Nulos ou NaN ****')
dfm = df['Massa\t'].isnull()
print(df[dfm])
print()

print('**** Colocando a média onde houver NaN ****')
media = df['Massa\t'].mean()

df.loc[dfm, 'Massa\t'] = media
print(df[dfm])
print()

# **** Transformando o Volume e Massa em inteiros ****
df = df.astype({"Volume": int, "Massa\t": int})

# **** Corrigindo o nome da coluna Massa ****
df.replace({'Massa\t':'Massa'},inplace=True)

print('*** Dados do DataFrame ***')
print('**************************')
print(f'Linhas e Colunas: {df.shape}')
print()
print(df.info())
print()
print(df.head())
print()
print(df.tail())
print()
```

```python
#Gravando o arquivo de Saída
df.to_csv (r'Volume_20200414.csv', index = False, header = False)
```
